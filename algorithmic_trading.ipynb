```
import pandas as pd
import talib
import numpy as np
```

```
df=pd.read_csv('BHARTIARTL.csv')
df.drop(columns=['date_id','timestamp','id','tradingSymbol'],inplace=True)
df
```

```
candle_names = talib.get_function_groups()['Pattern Recognition']
candle_names
important_candles=['CDLHAMMER','CDLINVERTEDHAMMER','CDLPIERCING','CDLENGULFING']
```

```
op = df['open']
hi = df['high']
lo = df['low']
cl = df['close']
```

```

for candle in important_candles:
    df[candle] = getattr(talib, candle)(op, hi, lo, cl)

```

```
updated_row=[]
for ind in df.index:
     if ind+1<df.shape[0]:
       updated_row.append((df.iloc[ind+1,1]-df.iloc[ind+1,0])/df.iloc[ind+1,0])

updated_row.append(0)
df['future_price']=updated_row
df
```

```
pattern_tick = df[df['CDLHAMMER'] > 0] 
print(pattern_tick.count())
positive_result = df[(df['CDLHAMMER'] > 0) & (df['future_price']>0)]
print(positive_result.count())
positive_result
```
